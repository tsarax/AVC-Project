```python
def list_creation_w(input1):
    trybeds = input1.copy()
    trybaths = input1.copy()
    trysqft = input1.copy()
    tryfloors= input1.copy()
    trywater = input1.copy()
    trycondition= input1.copy()
    trybasement = input1.copy()
    tryreno = input1.copy()

    if input1[0] >1: 
        trybeds[0] = trybeds[0]  - 1
    elif input1[0] == 1:
        trybeds[0] = trybeds[0]
           
    if input1[1] >1: 
        trybaths[1] = trybaths[1] - 1
    elif input1[1] == 1:
        trybaths[1] = trybaths[1]    
        
        
    trysqft[2]= trysqft[2] - trysqft[2]*.05

    if input1[3] > 1:
        tryfloors[3]= tryfloors[3] -1
    elif input1[3] == 1:
        tryfloors[3] = tryfloors[3]
    
    if input1[4] == 1:
        trywater[4]= trywater[4] -1
    elif input1[4] != 1:
        trywater[4] = trywater[4]
    
    if input1[5] >1:     
        trycondition[5]= trycondition[5] -1
    elif input[5] == 1:
        trycondition[5] = trycondition[5]
    
    if input1[6] == 1:
        trybasement[6]= trybasement[6] -1    
    elif input1[6] != 1:
        trybasement[6] = trybasement[6]
        
    if input1[8] == 1:
        tryreno[8]= tryreno[8]-1
    elif input1[8] != 1:
        tryreno[8] = tryreno[8]
    
    list1=[trybeds,trybaths,trysqft, tryfloors, trywater, trycondition, trybasement, tryreno]
    return list1
```

```python
def decreased_price_w(m, l):
    min_index=m
    list1=l
    picked_list = list1[min_index]
    item_list = []
    if min_index == 0:
        item_list.append('bedrooms')
        item_list.append(picked_list[0])
    if min_index == 1:
        item_list.append('bathrooms')
        item_list.append(picked_list[1])
    if min_index == 2:
        item_list.append('lot square footage')
        item_list.append(np.exp(picked_list[2]))
    if min_index == 3:
        item_list.append('floors')
        item_list.append(picked_list[3])
    if min_index == 4:
        item_list.append('waterfront view')
        item_list.append('none') 
    if min_index == 5:
        item_list.append('condition')
        item_list.append(picked_list[5])
    if min_index == 6:
        item_list.append('basement')
        item_list.append('none')  
    if min_index == 8:
        item_list.append('renovation')
        item_list.append('none')    
        
    return item_list    
```
